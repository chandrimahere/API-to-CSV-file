# Importing the required libraries

```python
import pandas as pd
import requests
import json
```

```python
url="https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1"
```

# Reading the data through API

```python
JSONContent = requests.get(url).json()
content = json.dumps(JSONContent, indent = 5, sort_keys=True)
print(content)
```

```python
df=[]
```

# Loop to extract the data from consecutive webpages

```python
for i in range(10):
    url = "https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page="
    JSONContent = requests.get(url+str(i+1)).json()
    result=JSONContent['results']
    for j in result:
        df.append([j['id'],j['title'],j['overview'],j['release_date'],j['popularity'],j['vote_average']])
```

# Converting the extracted data into a single dataframe

```python
data=pd.DataFrame(df)
```

# Naming the columns of the created dataframe

```python
data.columns=['Id','Title','Overview','Release_Date','Popularity','Vote_Average']
```
